## Career survey analysis: Data cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Analyzing_Career_Trends.csv')
df
```

```python
df.info()
```

```python
df.describe()

# most common functional area: IT & software
# perk that they want: salary
# most people do not want to work for a different industry (need to upskill)
# most people want to do an AI startup
```

```python
# drop college name and city since they are not relevant
df.head()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# drop empty columns

df_new = df.dropna(axis=1, how='all')
df_new
```

```python
# Fill NaN values in the specified column with "N/A not yet employed"
column_to_fill = 'If you are already an employee are you happy with your new job ?'
df_new[column_to_fill] = df_new[column_to_fill].fillna("N/A not yet employed")
df_new
```

```python
another_column = 'What are the Areas in which you did your internship ?'
df_new[another_column] = df_new[another_column].fillna("Unknown")
df_new
```

```python
df_new.drop(['Which City you would prefer to work in ?', 'How much time do you spend on your Mobile everyday ?'], axis=1, inplace=True)
```

```python
df_new
```

```python
last_to_fill = 'If yes, what area you would want a startup?'
channel_job = 'What channels would you use looking for Jobs ?'
df_new[last_to_fill] = df_new[last_to_fill].fillna("Do not wish to start a startup")
df_new[channel_job] = df_new[channel_job].fillna('Unknown')
df_new

# pretty clean now
```

```python
# drop all the rows that contain any 'NaN' values in them
# Sort the DataFrame so that all rows with NaN values are at the top
df_new_sorted = df_new.sort_values(by=df_new.columns.tolist(), ascending=[True]*len(df_new.columns), na_position='first')
# Display the sorted DataFrame to verify
df_new_sorted.head(20)  # Displaying more rows to show the NaN values at the top
```

```python
# Drop rows with any NaN values
df_cleaned = df_new_sorted.dropna()

# Display the cleaned DataFrame to verify
df_cleaned.head(20)
```

```python
df_cleaned.isnull().sum() #no more null!
```

```python
# save new df
df_cleaned.to_csv('cleaned_career_trends.csv')
```

```python
# Identify common functional areas & visualize them on barchart
common_functional_areas = career_survey['Which Functional Area do you want to work in ?'].value_counts().head(10)

plt.figure(figsize=(12, 8))
plt.barh(common_functional_areas.index, common_functional_areas.values, color='forestgreen')
plt.xlabel('Count')
plt.ylabel('Functional Area')
plt.title('Top 10 Preferred Functional Areas')
plt.gca().invert_yaxis()
plt.show()

# Lots of IT, software and suprisingly HR coming in at 2nd, 3rd, 4th
```

```python
# Identify most important factors for the first job
important_factors = career_survey['What is most important for you while opting for your first job?'].value_counts().head(10)

# Plotting most important factors for the first job
plt.figure(figsize=(12, 8))
plt.barh(important_factors.index, important_factors.values, color='firebrick')
plt.xlabel('Count')
plt.ylabel('Important Factors')
plt.title('Top 10 Important Factors for First Job')
plt.gca().invert_yaxis()
plt.show()

# top reasons: salary & chance to work on new projects, these factors over job role & company culture
```

```python
# Salary expectations - histogram
# Plotting the distribution of salary expectations
plt.figure(figsize=(12, 8))
sns.histplot(career_survey['How much salary would you expect if you are hired for a job?'], kde=True, color='black')
plt.xlabel('Expected Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salary Expectations')
plt.show()
```

```python
# Job preference by Gender
job_preferences_by_gender = career_survey.groupby(career_survey['Gender'])['Which Functional Area do you want to work in ?'].value_counts()
job_preferences_by_gender
```

```python
job_preferences_by_gender.T.plot(kind='bar', figsize=(14, 10), color=['navy', 'lime'])
plt.xlabel('Functional Area')
plt.ylabel('Count')
plt.title('Job Preferences by Gender')
plt.xticks(rotation=45)
plt.legend(['female','male'], title='Gender')
plt.show()

# no identifyable insights
```

```python
# happiness factors in a job

happiness_factors = career_survey['What factors make you Happy with a job?'].value_counts().head(10)
plt.figure(figsize=(12, 8))
plt.barh(happiness_factors.index, happiness_factors.values, color='coral')
plt.xlabel('Count')
plt.ylabel('Happiness Factors')
plt.title('Top 10 Happiness Factors for Your Job')
plt.gca().invert_yaxis()
plt.show()

# number one factor is growth opportunities
# high pay, recognition comes in next
```

```python
# Job demand trend based on internships
internship = career_survey['What are the Areas in which you did your internship ?'].value_counts().head(10)

internship.index

internship.values

fig, ax = plt.subplots()
ax.pie(internship.values, labels=internship.index, autopct='%1.1f%%',
       pctdistance=1.25, labeldistance=1.5)

# Data science & AI emerge at the top!
```

```python
# Top 10 areas people want to do a startup on

startup = career_survey['If yes, what area you would want a startup?'].value_counts().head(10)
plt.figure(figsize=(12, 8))
plt.barh(startup.index, startup.values, color='black')
plt.xlabel('Count')
plt.ylabel('Startup fields')
plt.title('Top 10 areas people want to do a startup on')
plt.gca().invert_yaxis()
plt.show()

# AI and EDUCATION!!!! - SKILLTECH :)
```

```python
import re
```

```python
# Function to clean and convert to numeric
def clean_and_convert(value):
    if isinstance(value, str):
        # Remove commas
        value = value.replace(',', '').strip()
        # Handle 'less than' or '-'
        if 'less than' in value:
            return float(re.sub(r'[^0-9.]', '', value.split('less than')[-1]))
        elif '-' in value:
            numbers = re.findall(r'[0-9]+', value)
            return sum(map(float, numbers)) / len(numbers)  # Take the average of the range
        elif value == '':
            return float('nan')  # Convert empty strings to NaN
        else:
            return float(re.sub(r'[^0-9.]', '', value))
    return value
```

```python
# apply the function to the column
career_survey['Monthly Average Family Income'] = career_survey['Monthly Average Family Income'].apply(clean_and_convert)
career_survey['Monthly Average Family Income']
```
